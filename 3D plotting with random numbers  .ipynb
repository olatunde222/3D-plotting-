**Importing the necessary libraries**


```python
import numpy as np # for random numbers generation 
from matplotlib import pyplot as plt # for plotting 
from mpl_toolkits import mplot3d 
```

**Single plotting**

```python
ax = plt.axes(projection="3d")
ax.scatter(3,5,7)
plt.show()
```

**Multiple plotting**

```python
ax = plt.axes(projection = "3d")

shape = int(500)

x_data = np.random.randint(0,50,(shape,))
y_data = np.random.randint(0,50,(shape,))
z_data = np.random.randint(0,50,(shape,))

ax.scatter(x_data,y_data,z_data,marker='v',color='k')
ax.set_title('multiple plotting')
ax.set_xlabel('x axis random data')
ax.set_ylabel('y axis random data')
ax.set_zlabel('z axis random data')
plt.show()
```

**plotting a function**

```python
ax = plt.axes(projection = "3d")
X = np.arange(0,50,0.1)
Y = np.arange(0,50,0.1)
Z = X * Y

ax.scatter(X,Y,Z)
ax.set_title('function plotting')
ax.set_xlabel('x axis random data')
ax.set_ylabel('y axis random data')
ax.set_zlabel('z axis random data')
plt.show()
```

```python
ax = plt.axes(projection = "3d")
X = np.arange(0,50,0.1)
Y = np.arange(0,50,0.1)
Z = np.sin(X) * np.cos(Y)

ax.plot(X,Y,Z)
ax.set_title('function plotting')
ax.set_xlabel('x axis random data')
ax.set_ylabel('y axis random data')
ax.set_zlabel(' sin and cosine ')
plt.show()
```

**meshgrid plotting**

```python
ax = plt.axes(projection = "3d")

X_data = np.arange(0,50,0.1)
Y_data = np.arange(0,50,0.1)
X,Y = np.meshgrid(X_data,Y_data)
Z = X * Y

ax.scatter(X,Y,Z)
ax.set_title('meshgrid plotting')
ax.set_xlabel('x axis random data')
ax.set_ylabel('y axis random data')
ax.set_zlabel('muliplication of x and y')
plt.show()
```

**Surface plotting**

```python
ax = plt.axes(projection = "3d")

X_data = np.arange(-5,5,0.1)
Y_data = np.arange(-5,5,0.1)
X,Y = np.meshgrid(X_data,Y_data)
Z = np.sin(X) * np.sin(Y)

ax.plot_surface(X,Y,Z)
ax.set_title('surface plotting')
ax.set_xlabel('x axis random data')
ax.set_ylabel('y axis random data')
ax.set_zlabel('sin and cosine of x&y')
plt.show()
```

```python
ax = plt.axes(projection = "3d")

X_data = np.arange(-5,5,0.1)
Y_data = np.arange(-5,5,0.1)
X,Y = np.meshgrid(X_data,Y_data)
Z = np.sin(X) * np.sin(Y)

ax.plot_surface(X,Y,Z,cmap='plasma')
ax.set_title('surface plotting')
ax.set_xlabel('x axis random data')
ax.set_ylabel('y axis random data')
ax.set_zlabel('sin and cosine of x&y')
plt.show()
```
